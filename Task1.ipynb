## Task 1: Predicting score based on hours of study using Linear Regression Model


### Importing Imp Library 

```python
# Import Imp Library
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Importing Data.csv

```python
# import Data.csv
#df=pd.read_csv('http://bit.ly/w-data')
df=pd.read_csv('Data.csv')
df.head()
```

### Ploting Graph of Hours Vs Score

```python
df.plot(x="Hours",y="Scores",style='o')
plt.grid()
plt.xlabel("Hours of study")
plt.ylabel("Score")
plt.title('Number of Hours of Study Vs Score')
```

### Seperating Data into (Input) and (OutPut)

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,1].values
```

### Randomly Splitting data into training set and testing set

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test=tts(X,y,test_size=0.2,random_state=0)
```

### Trining Linear Regression Model with X_train and y_train

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)

print("Training Complete!!!")
```

### Plot Trained Model with Given Data

```python
line=regressor.coef_*X + regressor.intercept_

plt.scatter(X,y)
plt.plot(X,line)
plt.grid()
plt.show()
```

### Making Prediction
#### Precicting output with trained model

```python
print(X_test)
y_pred=regressor.predict(X_test)
```

### comparing Actual and Predicted output

```python

ddf=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
ddf
```

### Puting Our Own Input

```python
hours=[[20]]
own_pred=regressor.predict(hours)
print("No of Hours : ",hours)
print("Pred val: ",own_pred)
```

# Evaluating The Model
### Taking mean of  difference of actual and predicted output

```python
from sklearn import metrics
print("Mean Absolute Error: ",metrics.mean_absolute_error(y_test,y_pred))
```
